### 2. 결측치 개수 확인

```python
df.isnull().sum()
```

### 3. 가사 EMPTY 찾아서 삭제

```python
condition = (df.가사=='EMPTY') # 조건식 작성

df[condition].info() 
```

```python
idx=df[df['가사']=='EMPTY'].index
df=df.drop(idx)
```

```python
df.info()
```

### 4. 가사 중복값 제거

```python
df.info()
```

```python
df=df.drop_duplicates(['가사']) # 가사가 같으면 중복이므로 제거.(inst 등 제거)
```

```python
df.info()
```

### 5. index 초기화 해주고 파일로 저장

```python
df=df.reset_index()
df=df.drop(['Unnamed: 0'],axis=1)
df=df.drop(['index'],axis=1)
```

```python
df
```

```python
df.to_csv("장범준_lyrics_clean.tsv", sep = "\t", encoding='cp949')
```
