```python
from bs4 import BeautifulSoup
from selenium import webdriver
driver = webdriver.Chrome('chromedriver.exe')
import pandas as pd
```

# 1페이지 (50곡)

```python
title_1=[]
songId_1=[]
lyrics_1=[]
LYRIC_1=[]
singer_1=[]

url="https://www.melon.com/search/song/index.htm?q=%EC%95%88%EC%98%88%EC%9D%80&section=&searchGnbYn=Y&kkoSpl=Y&kkoDpType=&mwkLogType=T#params%5Bq%5D=%25EC%2595%2588%25EC%2598%2588%25EC%259D%2580&params%5Bsort%5D=hit&params%5Bsection%5D=all&params%5BsectionId%5D=&params%5BgenreDir%5D=&po=pageObj&startIndex=1"
driver.get(url)

tmp=driver.find_elements_by_class_name('fc_gray')

for i in tmp:
    title_1.append(i.text)
    
# tmp=driver.find_elements_by_class_name('fc_gray')

for i in range(len(tmp)):
    if tmp[i].get_attribute('href')[-10] == ',':
        songId_1.append(tmp[i].get_attribute('href')[-9:-2])
    else:
        songId_1.append(tmp[i].get_attribute('href')[-10:-2])

tmp=driver.find_elements_by_id('artistName')
for i in tmp:
    singer_1.append(i.text)
        
for i in range(len(songId_1)):
    driver.get("https://www.melon.com/song/detail.htm?songId="+songId_1[i])
    tmp=driver.find_element_by_id('d_video_summary')
    lyrics_1.append(tmp.text)
    
    
for i in lyrics_1:
    LYRIC_1.append(i.replace("\n",""))
    
df_1=pd.DataFrame({"제목":title_1,"가수":singer_1,"가사":LYRIC_1})
```
